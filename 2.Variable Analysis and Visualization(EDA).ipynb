# Variable Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('Variable_final.csv')
```

```python
# Convert all columns to categorical type
df = df.astype("category")

print(df.dtypes)
```

## Univariate analysis

```python

categorical_vars = [
    'SEX', '_AGE_GROUP', '_RACEGR3', '_EDUCA_GROUP', '_INCOM_GROUP',
    '_SMOKER3_GROUP', '_BMI5CAT', '_FRTLT1A', '_VEGLT1A', '_TOTINDA',
    '_RFBING5', '_ECIGSTS_GROUP', '_PHYS14D', '_MENT14D', 'CHECKUP1_GROUP',
    '_LTASTH1'
]

# Calculate frequencies and percentages for each variable and visualize them
for var in categorical_vars:
    print("Variable:", var)
    freq = df[var].value_counts(dropna=False)      
    percent = df[var].value_counts(normalize=True, dropna=False) * 100
    print("Frequency:")
    print(freq)
    print("Percentage:")
    print(percent.round(2))
    print("\n-----------------\n")
    
    # Pie Chart
    plt.figure(figsize=(6, 6))
    plt.pie(freq, labels=freq.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'{var} Distribution(Pie)')
    plt.show()
```

## Bivariate relationship analysis

```python
import numpy as np
from scipy.stats import chi2_contingency
from statsmodels.graphics.mosaicplot import mosaic

def cramers_v(confusion_matrix):
    chi2, p, dof, expected = chi2_contingency(confusion_matrix)
    n = confusion_matrix.values.sum()  
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1) * (r-1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1))), chi2, p



dependent = '_LTASTH1'
independent_vars = [
    'SEX', '_AGE_GROUP', '_RACEGR3', '_EDUCA_GROUP', '_INCOM_GROUP',
    '_SMOKER3_GROUP', '_BMI5CAT', '_FRTLT1A', '_VEGLT1A', '_TOTINDA',
    '_RFBING5', '_ECIGSTS_GROUP', '_PHYS14D', '_MENT14D', 'CHECKUP1_GROUP'
]


for var in independent_vars:
    print(f'Analysis variables: {var} vs {dependent}')
    ct = pd.crosstab(df[var], df[dependent], dropna=False)
    print("Crosstab:")
    print(ct)
    
    #Chi-square test & Cramer's V
    chi2, p, dof, expected = chi2_contingency(ct)
    cv, _, _ = cramers_v(ct)
    print(f"Chi-square = {chi2:.2f}, p-value = {p:.4f}, Cramer's V = {cv:.4f}")
    
    # Visualization: Stacked Bar Chart
    plt.figure(figsize=(8, 4))
    sns.countplot(x=var, hue=dependent, data=df, order=ct.index)
    plt.title(f'{var} vs {dependent} - Stacked Bar Chart')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    
    # Visualization: Mosaic Plot
    plt.figure(figsize=(16, 10))
    mosaic(df, [var, dependent], title=f'{var} vs {dependent} - Mosaic diagram')
    plt.show()
    
    print("\n-----------------\n")


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2, p, dof, expected = chi2_contingency(confusion_matrix)

    n = confusion_matrix.values.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape

    if n - 1 == 0:
        return 0, chi2, p
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    denom_corr = min((kcorr - 1), (rcorr - 1))
    if denom_corr <= 0:
        return 0, chi2, p
    return np.sqrt(phi2corr / denom_corr), chi2, p

vars_list = [
    'SEX', '_AGE_GROUP', '_RACEGR3', '_EDUCA_GROUP', '_INCOM_GROUP',
    '_SMOKER3_GROUP', '_BMI5CAT', '_FRTLT1A', '_VEGLT1A', '_TOTINDA',
    '_RFBING5', '_ECIGSTS_GROUP', '_PHYS14D', '_MENT14D', 'CHECKUP1_GROUP', '_LTASTH1'
]

cramers_matrix = pd.DataFrame(index=vars_list, columns=vars_list, dtype=float)

# Compute Cramer's V between each pair of variables
for var1 in vars_list:
    for var2 in vars_list:
    
        ct = pd.crosstab(df[var1], df[var2], dropna=False)
        cv, chi2, p = cramers_v(ct)
        cramers_matrix.loc[var1, var2] = cv

# Draw a heat map
plt.figure(figsize=(12, 10))
sns.heatmap(cramers_matrix.astype(float), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Pairwise Cramer's V Heatmap")
plt.tight_layout()
plt.show()
```
